```python
# ==========================
# AI RESUME ANALYZER
# JUPYTER NOTEBOOK VERSION
# ==========================

import pdfplumber
import pandas as pd

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity


# ==========================
# SKILL DATABASE
# ==========================

SKILLS = [
    "python",
    "sql",
    "tableau",
    "power bi",
    "excel",
    "machine learning",
    "deep learning",
    "tensorflow",
    "keras",
    "pandas",
    "numpy",
    "scikit-learn",
    "nlp",
    "rag",
    "transformers",
    "aws",
    "azure",
    "docker",
    "git",
    "flask",
    "powerpoint",
    "statistics",
    "data analysis",
    "data visualization",
    "xgboost",
    "random forest",
    "logistic regression"
]


# ==========================
# LOAD EMBEDDING MODEL
# ==========================

print("Loading model...")

model = SentenceTransformer("all-MiniLM-L6-v2")

print("Model loaded successfully!")


# ==========================
# FUNCTIONS
# ==========================

def extract_pdf_text(pdf_path):

    text = ""

    with pdfplumber.open(pdf_path) as pdf:

        for page in pdf.pages:

            page_text = page.extract_text()

            if page_text:
                text += page_text + "\n"

    return text


def extract_skills(text):

    text = text.lower()

    found_skills = []

    for skill in SKILLS:

        if skill.lower() in text:
            found_skills.append(skill)

    return sorted(list(set(found_skills)))


def calculate_similarity(resume_text, jd_text):

    resume_embedding = model.encode([resume_text])

    jd_embedding = model.encode([jd_text])

    similarity_score = cosine_similarity(
        resume_embedding,
        jd_embedding
    )[0][0]

    return round(similarity_score * 100, 2)


def recruiter_feedback(score):

    if score >= 85:
        return "Excellent Match - Highly Recommended"

    elif score >= 70:
        return "Good Match - Recommended"

    elif score >= 50:
        return "Moderate Match - Needs Improvement"

    else:
        return "Low Match - Not Recommended"


# ==========================
# INPUTS
# ==========================

resume_path = r"C:\Users\KOWSIK.S\Resume\ATS\Kowsik_S_Resume_v6.pdf"
 

jd_text = """
Data Analyst

Location: Bangalore, India

Experience Required:
2 to 4 Years

Job Summary:

We are looking for a Data Analyst who can collect, clean, analyze, and interpret large datasets to support business decision-making. The candidate should be able to identify trends, generate insights, and create dashboards and reports for stakeholders.

Key Responsibilities:

- Collect, clean, and validate data from multiple sources.
- Perform exploratory data analysis (EDA) to identify trends and patterns.
- Develop dashboards and reports using Tableau or Power BI.
- Write complex SQL queries for data extraction and analysis.
- Work with stakeholders to understand business requirements.
- Create data visualizations and present actionable insights.
- Monitor key business metrics and performance indicators.
- Support predictive analytics and machine learning initiatives.
- Prepare documentation and maintain data quality standards.

Required Skills:

- Python
- SQL
- Pandas
- NumPy
- Statistics
- Data Analysis
- Data Visualization
- Tableau
- Power BI
- Microsoft Excel

Preferred Skills:

- Machine Learning
- Scikit-learn
- Deep Learning
- NLP
- TensorFlow
- Git
- AWS

Qualifications:

- Bachelor's degree in Engineering, Computer Science, Statistics, Mathematics, or related field.
- Strong analytical and problem-solving skills.
- Good communication and presentation skills.
- Ability to work independently and in a team environment.

Nice to Have:

- Experience with predictive modeling.
- Experience with dashboard development.
- Knowledge of business intelligence tools.
- Understanding of machine learning concepts.
"""


# ==========================
# ANALYSIS
# ==========================

print("\nReading Resume...")

resume_text = extract_pdf_text(resume_path)

resume_skills = extract_skills(resume_text)

jd_skills = extract_skills(jd_text)

score = calculate_similarity(
    resume_text,
    jd_text
)

matched_skills = list(
    set(resume_skills).intersection(jd_skills)
)

missing_skills = list(
    set(jd_skills).difference(resume_skills)
)


# ==========================
# RESULTS
# ==========================

print("\n" + "="*50)
print("AI RESUME ANALYSIS REPORT")
print("="*50)

print(f"\nMATCH SCORE : {score}%")

print("\nRECRUITER FEEDBACK:")
print(recruiter_feedback(score))

print("\nMATCHED SKILLS:")
for skill in sorted(matched_skills):
    print("✓", skill)

print("\nMISSING SKILLS:")
for skill in sorted(missing_skills):
    print("✗", skill)

print("\nTOTAL MATCHED SKILLS:", len(matched_skills))
print("TOTAL MISSING SKILLS:", len(missing_skills))


# ==========================
# DATAFRAME OUTPUT
# ==========================

results_df = pd.DataFrame({
    "Matched Skills": pd.Series(sorted(matched_skills)),
    "Missing Skills": pd.Series(sorted(missing_skills))
})

print("\nSkill Comparison Table:")
display(results_df)
```

```python
from flask import Flask, jsonify

app = Flask(__name__)

@app.route("/analyze")
def analyze():

    resume_text = extract_pdf_text(resume_path)

    resume_skills = extract_skills(resume_text)

    jd_skills = extract_skills(jd_text)

    score = calculate_similarity(
        resume_text,
        jd_text
    )

    matched_skills = list(
        set(resume_skills).intersection(jd_skills)
    )

    missing_skills = list(
        set(jd_skills).difference(resume_skills)
    )

    recommendation = recruiter_feedback(score)

    return jsonify({
        "Match Score": score,
        "Matched Skills": matched_skills,
        "Missing Skills": missing_skills,
        "Recommendation": recommendation
    })
```

```python
app.run(port=5000)
```
